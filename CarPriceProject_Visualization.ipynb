```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("car_project_cleaned.csv")
df.head()
```

```python
df.columns
```

```python
df["selling_price"].min() , df["selling_price"].max()
```

```python
plt.figure(figsize = (15,5))
sns.barplot(x= "year",y = "selling_price",data= df)
plt.title("Selling Price Every Year")
plt.show()
```

```python
sns.pairplot(data = df)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x="owner")
plt.title("Count of Cars by Ownership Type")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x="seats")
plt.title("Count of Cars by Number of Seats")
plt.yticks()
plt.show()
```

```python
plt.figure(figsize = (8,5))
sns.countplot(data=df, x='fuel', hue='transmission', palette='prism')
plt.show()
```

```python
plt.figure(figsize = (8,5))
sns.countplot(data=df, x="fuel", hue= "owner", palette="hot")
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.scatterplot(data=df, x='km_driven', y='mileage(km/ltr/kg)', hue='fuel', palette='viridis')
plt.title('Mileage vs Km Driven')
plt.xlabel('Km Driven')
plt.ylabel('Mileage (km/ltr/kg)')
plt.show()
```
